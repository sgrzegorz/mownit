```julia
strf = String[]
push!(strf,"sin(x)-x/2")
push!(strf,"2 * x - exp( - x )")
push!(strf," x * exp( - x )")
push!(strf,"cos ( x ) - x")
push!(strf,"( x + 3 ) * ( x - 1 )^2")
push!(strf,"x^3")
separator = "#########################################################################################"

f = Function[]
push!(f,x->sin(x)-x/2)
push!(f,x-> 2 * x - exp( - x ))
push!(f,x-> x * exp( - x ))
push!(f,x->cos( x ) - x)
push!(f,x->( x + 3 ) * ( x - 1 )^2)
push!(f,x->x^3)
println(f[1](0))

function print_charts(f)
    for i in 1:length(f)
        display(plot(f[i],-4,4,label=strf[i]))
        flush(stdout)
        println(separator)
                flush(stdout)

    end
end

print_charts(f)
```

```julia
#  bisection method
for i in 1:length(f)
    x = find_zero(f[i],(-4,4),verbose=true)
    println("f = ",strf[i],"\n is_zero(f(",x,") ", iszero(f[i](x)))
    println("f(",x,") = ",f[i](x))
    println(separator)
end

```

```julia
#  newton method
for i in 1:length(f)
    D(f) = x->ForwardDiff.derivative(f, float(x))
    x = find_zero((f[i], D(f[i])),0, Roots.Newton(),verbose=true)
    println("f = ",strf[i],"\n is_zero(f(",x,") ", iszero(f[i](x)))
    println("f(",x,") = ",f[i](x))
    println(separator)
end

```

```julia
#  steffensen method
for i in 1:length(f)
    x = find_zero(f[i], 0, Order2(), verbose=true) 
    println("f = ",strf[i],"\n is_zero(f(",x,") ", iszero(f[i](x)))
    println("f(",x,") = ",f[i](x))
    println(separator)
end
```

```julia
# 3B newton fractal

f = z->z^3-1
df = z->3*z^2




X = 1000
Y = 1000
T = zeros(RGB,Y,X)

# SCALING RANGES
# OldRange = (OldMax - OldMin)  
# NewRange = (NewMax - NewMin)  
# NewValue = (((OldValue - OldMin) * NewRange) / OldRange) + NewMin

function get_shade_of_colour(colour, iteration,MAX_ITERATION)
    
    old_range = MAX_ITERATION-1
    new_range = 1- (0)
    change = (iteration-1)*new_range/old_range  +0

    if colour == RGB(1,0,0)
        return RGB(1-change,0,0)
    elseif colour == RGB(0,1,0)
        return RGB(0,1-change,0)
            
    elseif colour == RGB(0,0,1)
        return RGB(0,0,1-change)
    end
end

function get_colour(y,x)
    old_range = X-1
    new_range = 1- (-2.5)
    zx = (x-1)*new_range/old_range  -2.5
    
    old_range = Y-1
    new_range = 1-(-1)
    zy =(y-1)*new_range/old_range -1
    
    z = zx +(zy)im
    roots = [1+0im, -0.5 + (sqrt(3)/2)im,-0.5 - (sqrt(3)/2)im]
    colours = [RGB(0,1,0),RGB(1,0,0),RGB(0,0,1)]   
    MAX_ITERATION = 20
    for iteration in 1:MAX_ITERATION
        
        z -= f(z)/df(z)
        tolerance = 0.000001
        for i in 1:length(roots)

            difference = z - roots[i]
            if abs(real(difference)) <tolerance && abs(imag(difference)) <tolerance
                return get_shade_of_colour(colours[i],iteration,MAX_ITERATION)
            end
        end
    end
    return RGB(0,0,0)
end



for x in 1:X
    for y in 1:Y
        T[y,x] = get_colour(y,x)
    end
end

display(T)


```
